```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy import stats
```

```python
### generate the data set of three columns, where each row is a triple of independent normal random variables
np.random.seed(999)
rows = 100
columns = 3
data = np.random.randn(rows, columns)
```

```python
### Define the function of constructing dataset according the conditions
def construct_dataset(a, b):
    df = pd.DataFrame(data)
    df.columns = ['z_i1', 'z_i2', 'z_i3']

    y_i1 = 1 + df['z_i1']
    y_i2 = 5 + 2 * df['z_i1'] + df['z_i2']
    u_i = a * (y_i1 - 1) + b * (y_i2 - 5) + df['z_i3']

    m_i2 = []
    for i in range(rows):
      if u_i[i] < 0:
        m_i2.append(1)
      else:
        m_i2.append(0)

    y_i2_miss = []
    for i in range(rows):
      if m_i2[i] == 1:
        y_i2_miss.append(np.nan)
      else:
        y_i2_miss.append(y_i2[i])

    df['y_i1'] = y_i1
    df['y_i2'] = y_i2
    df['u_i'] = u_i
    df['m_i2'] = m_i2
    df['y_i2_miss'] = y_i2_miss

    return(df)
```

#### Part(a)

We first extract the data that m_i2 = 0 and m_i2 = 1, seperately, and then we will draw plots for Y1 and Y2 under different cases

```python
### We first extract the data that m_i2 = 0 and m_i2 = 1, seperately
complete_case = df[df['m_i2'] == 0]
incomplete_case = df[df['m_i2'] == 1]
```

```python
## The plot for Y_1 under different cases
plt.figure(figsize = (10, 5))
plt.hist(complete_case['y_i1'], bins = 10, alpha = 0.7, label = 'Complete Y_1', color = 'blue')
plt.hist(incomplete_case['y_i1'], bins = 10, alpha = 0.7, label = 'Incomplete Y_1', color = 'green')
plt.title('Marginal Distribution of Y1')
plt.legend()
```

```python
## The plot for Y_2 under different cases
plt.figure(figsize = (10, 5))
plt.hist(complete_case['y_i2'], bins = 10, alpha = 0.7, label = 'Complete Y_2', color = 'blue')
plt.hist(incomplete_case['y_i2'], bins = 10, alpha = 0.7, label = 'Incomplete Y_2', color = 'green')
plt.title('Marginal Distribution of Y2')
plt.legend()
```

Notice that since we are under the case of a = 0 and b = 0, which means the missing mechanism for $y_{i2}$ only depends on the value of $z_{i3}$, but since $u$ is a latent variable, so this is MCAR, that the missing does not depend on both $Y_1$ and $Y_2$

#### Part(b)

Now we will present the two sample t-test for $Y_1$ in different group

```python
t_stat, p_value = stats.ttest_ind(complete_case['y_i1'], incomplete_case['y_i1'])
print("The t_stat is:", t_stat, "and the p_value is:", p_value)
```

I think that the above testing result shows that we cannot reject the null hypothesis that means of $Y_1$ is the same, which may indicate that the missing mechanism does not depend on $Y_1$. And we cannot make any statement about whether there is evidence of mechanism are not MCAR, MAR, and MNAR because we did not know how do $Y_2$ impact the missing mechanism.
